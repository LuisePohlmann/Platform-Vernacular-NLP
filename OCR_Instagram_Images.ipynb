```python
import pandas as pd 
import os
from PIL import Image 
from pytesseract import pytesseract 

file_names=[]
names=[]
dataframes=[]
place_holder= []

for root, dirs, files in os.walk(r'C:/Users/Lui/Desktop/Leuphana/01_Bachelor Arbeit/Data/alt/lvz.de'):
    for file in files:
        if file.endswith('.jpg'):
            file_names.append(os.path.join(root, file))

file_names=file_names[1:]
place_holder=list(range(len(file_names)))

for i in range(len(file_names)):
    path_to_tesseract = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
    image_path = file_names[i]
    img = Image.open(image_path) 
    pytesseract.tesseract_cmd = path_to_tesseract 
    text = pytesseract.image_to_string(img)
    dictionary = {place_holder[i]: [text, file_names[i]]}
    dataframes.append(dictionary)

dataframes2=[]
for i in range(len(dataframes)):
    new_df = pd.DataFrame.from_dict(dataframes[i], orient="index", columns=["Image Text", "Image Title"])
    dataframes2.append(new_df)

final_df = pd.concat(dataframes2, axis=0, join="outer", ignore_index=True)
final_df

for i in range(len(final_df["Image Title"])):
    x = final_df["Image Title"][i]
    if x[-5:-4]!= "C" and x[-5:-4]!= "1":
        final_df.drop(i, inplace=True)
        
final_df.to_csv("data_Instagram_with_image_text.csv",sep=";")
final_df
```
